```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='http://bit.ly/drinksbycountry'
```

```python
pd.read_table(url)
```

```python
data=pd.read_table(url,sep=',')
```

```python
data
```

```python
for i in ['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']:
    data[i]=data[i].fillna(data[i].median())
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['continent']=le.fit_transform(data['continent'])
data
```

```python
data['country']=le.fit_transform(data['country'])
data
```

```python
corr_matrix=data.corr()
```

```python
sns.heatmap(corr_matrix,vmin=-0.2,vmax=0.9,annot=True,cmap='YlGnBu')
plt.show()
```
